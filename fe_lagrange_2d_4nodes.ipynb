Definición de las variables de entorno

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
import sympy as sym
```

Definimos las funciones de Lagrange 1D

```python
def L0(x,x0,x1):
   return (x-x1)/(x0-x1)

def L1(x,x0,x1):
   return (x-x0)/(x1-x0)
```

Graficamos las funciones de lagrange para un caso particular

```python
x0Plt=0
x1Plt=1
xPlt = np.linspace(x0Plt, x1Plt, 10)

plt.plot(xPlt, L0(xPlt,x0Plt,x1Plt), color='red')
plt.plot(xPlt, L1(xPlt,x0Plt,x1Plt), color='green')

plt.show()
```

Ahora creamos la función de interpolación para el elemento 2D

```python
def N0(x,y,x0,y0,x1,y1):
  return L0(x,x0,x1)*L0(y,y0,y1)

def N1(x,y,x0,y0,x1,y1):
  return L1(x,x0,x1)*L0(y,y0,y1)

def N2(x,y,x0,y0,x1,y1):
  return L1(x,x0,x1)*L1(y,y0,y1)

def N3(x,y,x0,y0,x1,y1):
  return L0(x,x0,x1)*L1(y,y0,y1)
```

Graficamos las funciones de interpolación para verificar que estén correctas

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,subplot_kw={"projection": "3d"})

x0Plt=0
x1Plt=1
y0Plt=0
y1Plt=1
xPlt = np.linspace(x0Plt, x1Plt, 20)
yPlt = np.linspace(y0Plt, y1Plt, 20)
xPltSur, yPltSur = np.meshgrid(xPlt, xPlt)

# Plot the surface.
surf = ax1.plot_surface(xPltSur, yPltSur, N0(xPltSur,yPltSur,0,0,1,1),cmap=cm.coolwarm,linewidth=0, antialiased=False)
surf = ax2.plot_surface(xPltSur, yPltSur, N1(xPltSur,yPltSur,0,0,1,1),cmap=cm.coolwarm,linewidth=0, antialiased=False)
surf = ax3.plot_surface(xPltSur, yPltSur, N2(xPltSur,yPltSur,0,0,1,1),cmap=cm.coolwarm,linewidth=0, antialiased=False)
surf = ax4.plot_surface(xPltSur, yPltSur, N3(xPltSur,yPltSur,0,0,1,1),cmap=cm.coolwarm,linewidth=0, antialiased=False)
```
